```python
# from chart_studio.plotly import plot, iplot
# import chart_studio as go
# from chart_studio import tools
import pandas as pd
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.express as px
```

```python
df_white = pd.read_csv('../wine_quality/data/white.csv', delimiter=';')
df_red = pd.read_csv('../wine_quality/data/red.csv', delimiter=';')

def convert_class(x): return 'good' if x >= 7 else 'bad'
df_white['class'] = df_white['quality'].apply(convert_class)
df_red['class'] = df_red['quality'].apply(convert_class)
df_white['type'] = 'white'
df_red['type'] = 'red'
df = pd.concat([df_red, df_white], ignore_index=True)

```

```python

# Red wine :'fixed acidity', 'residual sugar' and 'alcohol' scores comparison with Scatter 3D Plot
fig = px.scatter_3d(df_red, 
                    x='alcohol', y='fixed acidity', z='residual sugar', color='class',
                   template='seaborn')
fig.show()
```

```python
# Red_wine: 2D - 2 features
px.scatter(df_red, x='alcohol', y='fixed acidity', color='class', 
           labels={'alcohol':'Alcohol', 'fixed acidity':'Fixed Acidity', 'class': 'Class'})
```

```python
# Red_wine: 2D - 2 features, with histogram and box
px.scatter(df_red, x='alcohol', y='fixed acidity', color='class', 
           marginal_x='histogram',
           marginal_y='box',
           labels={'alcohol':'Alcohol', 'fixed acidity':'Fixed Acidity', 'class': 'Class'})
```

```python
# Compare red-wine vs white-wine - 1 features
px.histogram(df, x="alcohol", color='class', 
            animation_frame='type',
            barmode = 'group',
            labels={'alcohol':'Alcohol', 'class':'Class'},
            )
```

```python
# Compare red-wine vs white-wine - 2 features
px.scatter(df, x="alcohol", y="fixed acidity", color='class', 
           animation_frame='type',
           labels={'alcohol':'Alcohol', 'fixed acidity':'Fixed Acidity', 'class':'Class'},
           template='plotly_dark')
```

```python
# Red_Wine : Quality and Alcohol
px.histogram(df, x="alcohol", color='quality', 
           animation_frame='quality',
           labels={'alcohol':'Alcohol', 'quality':'Quality'},
           template='seaborn')
```

```python
# Red-wine Visualize Distribution
fig = px.density_contour(df_red, x="alcohol", y="fixed acidity", color="class", marginal_x="rug", marginal_y="histogram")
fig.show()
```

```python
# Red_Wine Heatmap 2 features
fig = px.density_heatmap(df_red, x="alcohol", y="fixed acidity", marginal_x="rug", marginal_y="histogram")
fig.show()
```

```python
# Red Wine Scatter Matrix 
px.scatter_matrix(df_red,
                  dimensions=["alcohol", "fixed acidity", "residual sugar"], 
                  color="class",
                  template='seaborn')
                 
```
